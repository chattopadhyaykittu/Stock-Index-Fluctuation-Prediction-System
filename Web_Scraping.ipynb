```python
import requests
from bs4 import BeautifulSoup
```

```python
from datetime import date, datetime, timedelta

today = date.today()
```

```python
data = open('headlines_{}.html'.format(today),'r')
```

```python
soup = BeautifulSoup(data, 'html5lib')
```

```python
coverpage_news = soup.find_all('h2', class_='jsx-3979883281')
```

```python
coverpage_news
```

```python
span_news = soup.find_all('span', class_='jsx-3979883281')
```

```python
span_news
```

```python
len(coverpage_news)
```

```python
len(span_news)
```

```python
time_news = soup.find_all('time')
len(time_news)
```

```python
time_news
```

```python
today_news=[]
for itr in range(len(coverpage_news)):
    today_news.append([coverpage_news[itr].get_text(),time_news[itr].get_text()])
today_news
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
df = pd.DataFrame(today_news, columns = ['Headlines', 'Date'])
```

```python
df
```

```python

yesterday = datetime.now() - timedelta(1)
d1 = today.strftime("%B %d, %Y")
d1
```

```python
d2 = yesterday.strftime("%B %d, %Y")
d2
```

```python
for itr in range(3,len(d1)):
    if d1[itr]==" ":
        index=itr
        break
today_date = d1[0:3]+d1[index:len(d1)]
today_date
```

```python
for itr in range(3,len(d1)):
    if d2[itr]==" ":
        index=itr
        break
yesterday_date = d2[0:3]+d2[index:len(d1)]
yesterday_date
```

```python
df1 = df[df['Date'].str.contains(today_date)]
df1
```

```python
df2 = df[df['Date'].str.contains(yesterday_date)]
df2
```

```python
df3 = df1.append(df2)
df3
```
